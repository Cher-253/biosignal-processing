```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file = 'data/pilot/kevin.csv'
df = pd.read_csv(file)
df
```

```python
df['Duration'] = (pd.to_datetime(df['End']) - pd.to_datetime(df['Begin'])).dt.total_seconds()
```

```python
df
```

```python
import scipy.stats, json
from scipy import stats
def to_code(w):
    return "-".join(w.split("\\")[1:]).replace(' ', '_')

codes = df['Code'].unique()
for code in range(len(codes)):
    codes[code] = to_code(codes[code])
    


n = len(codes)
print "Constructing Count Matrix", "n=", n
count_matrix = np.ones((n, n), dtype=float)*0.00001

codes = codes.tolist()
for i in range(1, len(df['Code'])):
    try:
        previous = codes.index(to_code(df['Code'][i-1]))
        current = codes.index(to_code(df['Code'][i]))
        count_matrix[previous][current] += 1 
    except IndexError:
        pass
    except ValueError:
        pass
        
for i in range(n):
    de = codes[i]
    for j in range(n):
        count = count_matrix[i, j]
        if count == 0: continue
        para = codes[j]
#         print de, "["+str(count)+"]", para
count_matrix /= (count_matrix.sum(axis=1, keepdims=True))
count_matrix[count_matrix<0.01] = 0
count_matrix[count_matrix>0.99] = 1
plt.rcParams['figure.figsize'] = (10, 10)
im = plt.imshow(count_matrix, cmap='gray')
ax= plt.xticks(range(len(codes)), codes, fontsize=12,rotation=90)
plt.yticks(range(len(codes)), codes, fontsize=12)
ax = plt.gca()
ax.xaxis.tick_top()
grid = count_matrix.round(2)
for (j,i),label in np.ndenumerate(grid):
    ax.text(i,j,label,ha='center',va='center')
plt.show()
```

## Wrangling Data to Visualize as a Force Directed Graph

```python
actors = list(set(map(lambda x: x.split('-')[0], codes)))

nodes = []
for code in codes:
    d = code.split("-")
    actor = d[0]
    group = actors.index(actor)
    state = code
    nodes.append({'id': state, 'group': group})

links = []
for i in range(len(count_matrix)):
    for j in range(len(count_matrix)):
        source = codes[i]
        target = codes[j]
        value = count_matrix[i][j]
        if value == 0: continue
        else: links.append({'source': source, 'target': target, 'value': int(value * 10)})

duration_data = zip(df['Code'], df['Duration'])
duration = np.ones(len(codes), dtype=float) * 5.0
for c, dur in duration_data:
    code_i =  codes.index(to_code(c))
    duration[code_i] += float(dur)

duration = duration/np.max(duration)

for i in range(len(nodes)):
    node = nodes[i]
    node['size'] = duration[i] * 10 +5



data = {
    'nodes': nodes,
    'links': links
}
print json.dumps(data)

```
